````python
class Solution:
    def getLeastNumbers(self,arr,k):
        heap = []
        for a in arr:
            heappush(heap,a)
            if len(heap)>k:
                heappop(heap)
        return heap
````

````python
class Solution:
    def getLeastNumbers(self,nums,k):
        def partition(pi,L,R):
            nums[pi],nums[R] = nums[R],nums[pi]
            l,r = L,R-1
            while l<r:
                while l<r and nums[l]<=nums[pi]:
                    l += 1
                while l<r and nums[r]>=nums[pi]:
                    r -= 1
                nums[l],nums[r] = nums[r],nums[l]
            nums[l],nums[R] = nums[R],nums[l]
            return l
        
        L,R = 0, len(nums)-1
        ok = R+1
        while ok+1!=k:
            if ok+1<k:
                L = ok+1
            else:
                R = ok-1
            import random
            ok = partition(random.randint(L,R),L,R)
            print(ok,nums[:ok],nums[ok],nums[ok+1:],k)
        return nums[ok]
````
