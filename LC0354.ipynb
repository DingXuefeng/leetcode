# Russian dossel

Given a[i] and b[i], find largest set (i) such that

```
a[i]<=a[j] b[i]<=b[j] if i<j
```

```
1<=a,b.lengths<=5000
1<=a[i],b[i]<=10**4
```

````python
class Solution:
    def maxEnvelopes(self,envelopes):
        es = sorted(envelopes)
        print(es)
        n = len(es)
        dp = [0]*(n+1)
        j = 0
        ans = 0
        stack1 = [float('-inf')]
        stack3 = [0]
        stack2 = [float('inf')]
        stack4 = [0]
        stack5 = [[]]
        for i in range(n):
            while es[j][0]<es[i][0]:
                while stack1[-1]>es[j][1]:
                    stack1.pop()
                    stack3.pop()
                stack1.append(es[j][1])
                stack3.append(j)
                coll = []
                while stack2[-1]<es[j][1]:
                    stack2.pop()
                    stack4.pop()
                    coll = stack5.pop()+coll
                stack2.append(es[j][1])
                stack4.append(j)
                coll.append(stack2[-1])
                stack5.append(coll)
                j += 1
            # [0:j] are all ev with w less than i
            dp[i] = 1
            
            print(es[i][1],f'{str(stack1):40s}',f'{str(stack2):50s}',f'{str([es[k][1] for k in range(j) if es[k][1]<es[i][1]]):60s}', f'{str(stack5[-1]):40s}',max([-999]+[dp[k] for k in stack3]),max([dp[k] for k in stack4]),list(zip(*es))[1][:j],)
            print(stack5)
            for k in range(j):
                if es[k][1]<es[i][1]:
                    dp[i] = max(dp[i],dp[k]+1)
            ans = max(ans,dp[i])
        return ans
````

````python
class Solution:
    def maxEnvelopes(self,es):
        es = sorted(es)
        stack = []
        j = 0
        ans = 0
        for i,(w,h) in enumerate(es):
            while es[j][0]<w:
                if not stack or es[j][1]>=stack[-1]:
                    stack.append(es[j][1])
                else:
                    # stack[pos-1]<h, stack[pos]>=h
                    stack[bisect_left(stack,es[j][1])] = es[j][1]
                j += 1
            ans = max(ans,bisect_left(stack,h)+1)
        return ans
            
````
