# peak in a map

Given a matrix isWater[i][j] (0 or 1), find a matrix a with max maximum.
a[i][j] = 0 if isWater[i][j] = 0
maximum diff between a[i][j] and its neighbour is 1

```
1<=m,n<=1000
at least 1 0 in isWater
```

## Idea

greedy. pick grid[x][y] = max(grid[x][y])
use bfs

````python
class Solution:
    def heighestPeak(self,isWater):
        rows,cols = len(isWater),len(isWater[0])
        q = [(i,j) for i in range(rows) for j in range(cols) if isWater[i][j]==1]
        h = 0
        ans = [[0]*cols for _ in range(rows)]
        while q:
            for i,j in q:
                ans[i][j] = h
            q = set([(i2,j2) for i,j in q for i2,j2 in [(i-1,j),(i+1,j),(i,j-1),(i,j+1)] if 0<=i2<rows and 0<=j2<cols and ans[i2][j2]==0 and isWater[i2][j2]==0])
            h += 1
        return ans
````

**mistake**
- q = set(...)
- need 2 matrix, otherwise water is set to 0 in the first step.
- isWater==1 means it's water
