# Maximum XOR of two numbers in an array

constraints
```
1<=nums.length<=2*10**5
0<=nums[i]<=2**31-1
```

````python
## Analysis
0b0000000000000000000000000000000<=nums[i]<=0b1111111111111111111111111111111 (31x1 and 31x0)
if a^b = 0bx..x10..0, 0xi, then it will be larger 
````

````python
import numpy as np
nums = np.random.randint(0,(1<<31)-1,10)
# nums = [100,100,100]
ans,mask = 0,0
how = [0,None]
for i in range(31,-1,-1):
    mask = mask | (1<<i)
    print(bin(mask),mask)
    aset = {n&mask for n in nums}
    print([bin(x) for x in aset])
    tmp = ans|1<<i
    old,new = bin(ans),bin(tmp)
    print(f'{old} -> {new} ok?')
    for a in aset:
        if tmp^a in aset: # a^(tmp^a) = a^(a^tmp) = (a^a)^tmp = 0^tmp = tmp
            ans = tmp
            how = [a,tmp^a]
            break
    print(f'{old} -> {new} {f"ok +{how}" if ans==tmp else "not ok"}')
    print('-----')
````
